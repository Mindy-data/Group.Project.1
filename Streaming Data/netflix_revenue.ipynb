```python
#Dependencies & setup
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

```python
# Files to load
netflix_revenue = ("Streaming Data/netflix_revenue_updated.csv")
```

```python
# Load Netflix revenue data intoa a DataFrame
df_netflix_revenue = pd.read_csv("netflix_revenue_updated.csv")

df_netflix_revenue.head()
```

```python
df_netflix_revenue.info()
```

```python
x = df_netflix_revenue["Date"]
x
```

```python
# trim and remove unnecessary spacing 
def clean_str(col:str):
    if col[-1] == ' ':
        return clean_str(col[:-1])
    elif col[0] == ' ':
        return clean_str(col[1:])
    return col.replace('  ',' ')

df_netflix_revenue.columns = [clean_str(x) for x in df_netflix_revenue]
df_netflix_revenue.head()
```

```python
#x = df_netflix_revenue["Date"]
#y = df_netflix_revenue["Netflix Streaming Memberships"]

df_netflix_revenue.plot(x='Date', y='Netflix Streaming Memberships',
                        kind="bar", color="b",legend=False)
plt.title("Netflix Total Membership per Quarter")
plt.xlabel("Date")
plt.ylabel("Netflix Streaming Memberships")

plt.savefig("Output Data/NetflixMembership.png" , bbox_inches="tight")

plt.show()
```

```python
ax = df_netflix_revenue.plot(x='Date', y='Global Revenue',
                        kind="bar", color="b",legend=False, width=0.5, align="center")

def format_dollars(value, _):
    return f"${value:.0f}"

ax.yaxis.set_major_formatter(FuncFormatter(format_dollars))

plt.title("Netflix Revenue per Quarter")
plt.xlabel("Date")
plt.ylabel("Global Revenue")

plt.savefig("Output Data/NetflixRevenue.png", bbox_inches="tight")
plt.show()
```
