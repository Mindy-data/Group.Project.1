```python
import pandas as pd
import matplotlib.pyplot as plt

amc_stock= pd.read_csv('Stock_files/AMC.csv')
amc_stock['Ticker'] = 'AMC'
amc_stock.set_index('Ticker', inplace=True)
imax_stock= pd.read_csv('Stock_files/IMAX.csv')
imax_stock['Ticker'] = 'IMAX'
imax_stock.set_index('Ticker', inplace=True)
nflx_stock= pd.read_csv('Stock_files/NFLX.csv')
nflx_stock['Ticker'] = 'NFLX'
nflx_stock.set_index('Ticker', inplace=True)
```

```python
amc_stock.plot('Date','High')
plt.xticks(rotation = 50)
plt.title('AMC Stock')
plt.ylabel('Highest stock price')
plt.savefig('Stock_files/AMC_plot')
plt.show()
```

```python
nflx_stock.plot('Date','High')
plt.xticks(rotation = 50)
plt.title('NETFLIX Stock')
plt.ylabel('Highest stock price')
plt.savefig('Stock_files/nflx_plot')
plt.show()
```

```python
imax_stock.plot('Date','High')
plt.xticks(rotation = 50)
plt.title('IMAX Stock')
plt.ylabel('Highest stock price')
plt.savefig('Stock_files/imax_plot')
plt.show()
```
