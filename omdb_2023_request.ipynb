```python
import requests
import locale
locale.setlocale( locale.LC_ALL, '' )
import pandas as pd
from config import tmdb_key
base_params = dict(api_key=tmdb_key)

listing_params = dict(primary_release_year=2019, sort_by='revenue.desc')
listing_params.update(base_params)
response = requests.get('https://api.themoviedb.org/3/discover/movie', params=listing_params)
imdb_films = response.json()['results']

rows = []

film_params = dict(language='en-US',primary_release_year=2019)
film_params.update(base_params)
for film in imdb_films:
   # print(film['title'])
   film_revenue = requests.get('https://api.themoviedb.org/3/movie/{}'.format(film['id']), params=film_params)
   film_revenue = film_revenue.json()
   #print(locale.currency(film_revenue['revenue'], grouping=True ))
   row = dict(title=film['title'], revenue=film_revenue['revenue'])
   rows.append(row)


df_2019 = pd.DataFrame(rows, columns=['title', 'revenue'])
df_2019
```
